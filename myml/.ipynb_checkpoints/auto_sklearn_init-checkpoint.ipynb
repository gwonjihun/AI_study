```python
init_data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/datasets/train.csv")
```

```python
keys = ['Temperature', 'Fuel_Price', 'Promotion1',
       'Promotion2', 'Promotion3', 'Promotion4', 'Promotion5', 'Unemployment',
       'IsHoliday']
```

```python
fill_zero_data = init_data.copy()
for a in keys:
    fill_zero_data[a] = init_data[a].fillna(0)
```

```python
fill_zero_data
```

```python
fill_zero_data= fill_zero_data.drop(["id"],axis=1)
```

```python
fill_zero_data
```

```python
def get_month(date):
    month = date[3:5]
    month = int(month)
    return month

def get_day(date):
    day = date[0:2]
    month =  int(day)
    return month

def get_year(date):
    day = date[6:]
    day = int(day)
    return day

def holiday_to_number(isholiday):
    if isholiday == True:
        number = 1
    else:
        number = 0
    return number
```

```python
train_data= fill_zero_data[['Store','Temperature','Fuel_Price','Promotion1','Promotion2','Promotion3','Promotion4','Promotion5','Unemployment','IsHoliday','Weekly_Sales']]
```

```python

train_data['IsHoliday'] = fill_zero_data['IsHoliday'].apply(holiday_to_number)
```

```python
train_data['Month'] = fill_zero_data['Date'].apply(get_month)
train_data['Day'] = fill_zero_data['Date'].apply(get_day)
train_data['Year'] = fill_zero_data['Date'].apply(get_year)
```

```python
train_data
```

```python
dummy = pd.get_dummies(train_data['Store'],prefix='Store')
```

```python
train_data = pd.concat([train_data,dummy],axis=1)
```

```python
train_data = train_data.drop(['Store'],axis=1)
```

```python
train_data.columns
```

```python
X_train = train_data[['Temperature', 'Fuel_Price', 'Promotion1', 'Promotion2', 'Promotion3',
       'Promotion4', 'Promotion5', 'Unemployment', 'IsHoliday', 
       'Month', 'Day', 'Year', 'Store_1', 'Store_2', 'Store_3', 'Store_4',
       'Store_5', 'Store_6', 'Store_7', 'Store_8', 'Store_9', 'Store_10',
       'Store_11', 'Store_12', 'Store_13', 'Store_14', 'Store_15', 'Store_16',
       'Store_17', 'Store_18', 'Store_19', 'Store_20', 'Store_21', 'Store_22',
       'Store_23', 'Store_24', 'Store_25', 'Store_26', 'Store_27', 'Store_28',
       'Store_29', 'Store_30', 'Store_31', 'Store_32', 'Store_33', 'Store_34',
       'Store_35', 'Store_36', 'Store_37', 'Store_38', 'Store_39', 'Store_40',
       'Store_41', 'Store_42', 'Store_43', 'Store_44', 'Store_45']]
y_train = train_data[['Weekly_Sales']]
```

```python
import autosklearn.regression
```

```python
y_train = y_train.to_numpy()
```

```python
automl = autosklearn.regression.AutoSklearnRegressor(
    time_left_for_this_task=3600*6,per_run_time_limit=300,
    seed=10
)
```

```python
automl.fit(X_train,y_train)
```

```python
automl.get_models_with_weights()
```

```python
print(automl.leaderboard())
```
