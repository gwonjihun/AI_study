```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
wh_wine = pd.read_csv("datasets/winequality-red.csv",delimiter=";")
```

```python
rd_wine['target']=0
wh_wine['target']=1
```

```python
merge_data = pd.concat([wh_wine,rd_wine],axis=0)
```

```python
merge_data
```

```python
X = merge_data.iloc[:,:12]
```

```python
X
```

```python
y= merge_data.iloc[:,-1]
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,shuffle=True, random_state=2022, stratify=y)
```

```python
X_train
```

```python
sc = StandardScaler()
sc.fit(X_train)
```

```python
X_std_train = sc.transform(X_train)
```

```python
from tqdm import tqdm
from sklearn.model_selection import cross_val_score

hist = []

for k in tqdm(range(2,100)):
    model = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(model, X_std_train, y_train, cv=10, scoring="accuracy")

    hist.append(scores.mean())
```

```python
dicts = {}
for a,k in enumerate(hist):
    dicts[a] = k
```
