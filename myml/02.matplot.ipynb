```python
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
```

```python
xarr = np.random.rand(15).cumsum()
yarr = np.random.rand(15).cumsum()
```

```python
plt.figure(figsize=(2,2))
plt.plot(xarr)
plt.show()
```

```python
plt.figure()
plt.show()
# object만 만들어진 상황으로 그래프가 출력되지 않는다
```

```python
plt.figure()
plt.subplot() # 내부의 축을 만든다
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(12,12,144) # 내부의 축을 만든다
ax2 = fig.add_subplot(12,12,1)
ax1.plot(xarr)
ax2.plot(xarr)
plt.show()
```

```python
fig, axes = plt.subplots(2,2)
axes[0,1].plot(xarr)
axes[0,0].plot(yarr)
plt.show()
```

```python
xarr = np.random.rand(15).cumsum()
yarr = np.random.rand(15).cumsum()
zarr = np.random.rand(15).cumsum()
```

```python
plt.figure(figsize=(20,10))
```

```python
plt.plot(xarr, label='ko',color="#00D0F0", linestyle='-.', marker = "*")
plt.plot(yarr, label='us',color="coral")
plt.plot(zarr, label='jp')
plt.title("this is title")
plt.xlabel("this is xlabel")
plt.ylabel("this is ylabel")
plt.legend()
plt.show()
```

## Boxplot

```python
x = np.random.normal(0, 2.0, 1000)
y = np.random.normal(-3, 1.5, 500)
z = np.random.normal(1.2, 1.5, 1000)
```

```python
plt.boxplot([x,y,z])
plt.show()
```
