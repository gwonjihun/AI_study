```python
init_data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/datasets/train.csv")
```

```python
keys = ['Temperature', 'Fuel_Price', 'Promotion1',
       'Promotion2', 'Promotion3', 'Promotion4', 'Promotion5', 'Unemployment',
       'IsHoliday']
```

```python
fill_zero_data = init_data.copy()
for a in keys:
    fill_zero_data[a] = init_data[a].fillna(0)
```

```python
fill_zero_data
```

```python
fill_zero_data= fill_zero_data.drop(["id"],axis=1)
```

```python
fill_zero_data
```

```python
def get_month(date):
    month = date[3:5]
    month = int(month)
    return month

def get_day(date):
    day = date[0:2]
    month =  int(day)
    return month

def get_year(date):
    day = date[6:]
    day = int(day)
    return day

def holiday_to_number(isholiday):
    if isholiday == True:
        number = 1
    else:
        number = 0
    return number
```

```python
train_data= fill_zero_data[['Store','Temperature','Fuel_Price','Promotion1','Promotion2','Promotion3','Promotion4','Promotion5','Unemployment','IsHoliday','Weekly_Sales']]
```

```python

train_data['IsHoliday'] = fill_zero_data['IsHoliday'].apply(holiday_to_number)
```

```python
train_data['Month'] = fill_zero_data['Date'].apply(get_month)
train_data['Day'] = fill_zero_data['Date'].apply(get_day)
train_data['Year'] = fill_zero_data['Date'].apply(get_year)
```

```python
train_data
```

```python
dummy = pd.get_dummies(train_data['Store'],prefix='Store')
```

```python
train_data = pd.concat([train_data,dummy],axis=1)
```

```python
train_data = train_data.drop(['Store'],axis=1)
```
